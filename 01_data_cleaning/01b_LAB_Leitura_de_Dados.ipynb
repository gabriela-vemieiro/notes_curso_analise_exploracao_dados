```python
# Crie uma variável, `path`, contendo o caminho para o `baseball.db` contido em `sources/`
path = 'baseball.db'

# Crie uma conexão, `con`, que esteja conectada ao banco de dados em `path`
con = sq3.Connection(path)

# Crie uma variável, `query`, contendo uma consulta SQL que lê todos os dados da tabela

query = """
SELECT *
    FROM allstarfull
    ;
"""

allstar_observations = pd.read_sql(query, con)

# print(allstar_observations)

# Crie uma variável, tables, que lê todos os dados da tabela sqlite_master
all_tables = pd.read_sql('SELECT * FROM sqlite_master', con)
print(all_tables)

# Finja que você está interessado em criar um novo hall da fama do beisebol. Una e analise as tabelas para avaliar os 3 melhores jogadores de beisebol de todos os tempos
best_query = """
SELECT playerID, sum(GP) AS num_games_played, AVG(startingPos) AS avg_starting_position
    FROM allstarfull
    GROUP BY playerID
    ORDER BY num_games_played DESC, avg_starting_position ASC
    LIMIT 3
"""
best = pd.read_sql(best_query, con)
print(best.head())
```
